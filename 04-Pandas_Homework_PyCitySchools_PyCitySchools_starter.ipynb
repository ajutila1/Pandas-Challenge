```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(20)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate total number of schools
total_schools = len(school_data_complete["school_name"].unique())

#Calculate total number of students
total_students = school_data['size'].sum()

#Calculate total budget
budget = school_data['budget'].sum()
total_budget = "${:,.2f}".format(budget)

#Calculate averages of math and reading scores
average_math = school_data_complete['math_score'].mean()
average_reading = school_data_complete['reading_score'].mean()

#Calculate percent passing of math and reading
passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
total_passing_math = len(passing_math)
percent_passing_math = ((total_passing_math / len(school_data_complete['math_score'])) * 100)

passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
total_passing_reading = len(passing_reading)
percent_passing_reading = ((total_passing_reading / len(school_data_complete['reading_score'])) * 100)

#Calculate % passing both reading and math at same time
passing_both = len(school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)])
total_passing_both = ((passing_both / total_students) * 100)

#Create new dataframe to hold calculated data
summary_df = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students], "Total Budget": [total_budget],
                          "Average Math Score": [average_math], "Average Reading Score": [average_reading],
                          "% Passing Math": [percent_passing_math], "% Passing Reading": [percent_passing_reading],
                          "% Overall Passing": [total_passing_both]})

#Clone dataframe to format for clean presentation, retaining original dataframe in case of future calls
summary_formatted = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students], "Total Budget": [total_budget],
                          "Average Math Score": [average_math], "Average Reading Score": [average_reading],
                          "% Passing Math": [percent_passing_math], "% Passing Reading": [percent_passing_reading],
                          "% Overall Passing": [total_passing_both]})

summary_formatted["Average Math Score"] = summary_formatted["Average Math Score"].map("{:.2f}".format)
summary_formatted["Average Reading Score"] = summary_formatted["Average Reading Score"].map("{:.2f}".format)
summary_formatted["% Passing Math"] = summary_formatted["% Passing Math"].map("{:.2f}".format)
summary_formatted["% Passing Reading"] = summary_formatted["% Passing Reading"].map("{:.2f}".format)
summary_formatted["% Overall Passing"] = summary_formatted["% Overall Passing"].map("{:.2f}".format)

summary_formatted
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Create new table using merged dataframes
overview_table = school_data_complete[['school_name', 'reading_score', 'math_score', 'type', 'size', 'budget']]

#Group data by school name
school_group = overview_table.groupby(['school_name'])

#Pull type, size, and budget values per school
school_type = school_group['type'].unique()
school_size = school_group['size'].unique()
school_budget = school_group['budget'].unique()
school_budget = school_budget.astype(int)

#Calculate per student budget
per_student = (school_budget) / (school_size)
per_student = per_student.astype(int)

#Calculate average math, reading, and percent passing of each
average_math = school_group['math_score'].mean()

average_reading = school_group['reading_score'].mean()

students_by_school = school_group['math_score'].count()
math_passed_percent = (school_data_complete.loc[school_data_complete["math_score"] >= 70].groupby("school_name")["math_score"].count()/students_by_school).tolist()

reading_passed_percent = (school_data_complete.loc[school_data_complete["reading_score"] >= 70].groupby("school_name")["math_score"].count()/students_by_school).tolist()

#Calculate overall passing
overall_passing_percent = ((school_data_complete.loc[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name")["Student ID"].count())/students_by_school).tolist()

#Create new dataframe for overview data
overview_df = pd.DataFrame({"Type": school_type, "Total Students": school_size, "Total School Budget": school_budget,
                           "Per Student Budget": per_student, "Average Math Score": average_math, "Average Reading Score": average_reading,
                           "% Passing Math": math_passed_percent, "% Passing Reading": reading_passed_percent, "% Overall Passing": overall_passing_percent})

#Clone dataframe to format for clean presentation, retaining original dataframe in case of future calls
overview_formatted = pd.DataFrame({"Type": school_type, "Total Students": school_size, "Total School Budget": school_budget,
                           "Per Student Budget": per_student, "Average Math Score": average_math, "Average Reading Score": average_reading,
                           "% Passing Math": math_passed_percent, "% Passing Reading": reading_passed_percent, "% Overall Passing": overall_passing_percent})

overview_formatted["Total Students"] = overview_formatted["Total Students"].astype(int)
overview_formatted["Total School Budget"] = overview_formatted["Total School Budget"].map("${:,.2f}".format)
overview_formatted["Per Student Budget"] = overview_formatted["Per Student Budget"].map("${:.2f}".format)
overview_formatted["Average Math Score"] = overview_formatted["Average Math Score"].map("{:.2f}".format)
overview_formatted["Average Reading Score"] = overview_formatted["Average Reading Score"].map("{:.2f}".format)
overview_formatted["% Passing Math"] = (overview_formatted["% Passing Math"] * 100).map("{:.2f}".format)
overview_formatted["% Passing Reading"] = (overview_formatted["% Passing Reading"] * 100).map("{:.2f}".format)                                   
overview_formatted["% Overall Passing"] = (overview_formatted["% Overall Passing"] * 100).map("{:.2f}".format)

overview_formatted
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#Use sort function to show top five performing schools
overview_formatted = overview_formatted.sort_values("% Overall Passing", ascending=False)
overview_formatted.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create new table and group it by school name
math_by_grade_table = school_data_complete[['school_name', "grade", "math_score"]]
math_by_grade = math_by_grade_table.groupby(['school_name'])

#Calculate average math score per grade
ninth_grade_m = school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby("school_name")["math_score"].mean()
tenth_grade_m = school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby("school_name")["math_score"].mean()
eleventh_grade_m = school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby("school_name")["math_score"].mean()
twelfth_grade_m = school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby("school_name")["math_score"].mean()

#Create new dataframe using calculations
math_by_grade_df = pd.DataFrame({"9th": ninth_grade_m, "10th": tenth_grade_m, "11th": eleventh_grade_m, "12th": twelfth_grade_m})

#Format new dataframe
math_by_grade_df["9th"] = math_by_grade_df["9th"].map("{:.2f}".format)
math_by_grade_df["10th"] = math_by_grade_df["10th"].map("{:.2f}".format)
math_by_grade_df["11th"] = math_by_grade_df["11th"].map("{:.2f}".format)
math_by_grade_df["12th"] = math_by_grade_df["12th"].map("{:.2f}".format)

math_by_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Create bins using increments of 25 (Personal choice)
bins = [0, 590, 615, 640, 700]
school_spending_labels = ["<590", "591-615", "616-640", ">641"]

#Create reduced table using only needed columns
reduced_overview_spending = overview_df[["Per Student Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]

#Arrange data by bins
reduced_overview_spending["Spending Ranges (Per Student)"] = pd.cut(reduced_overview_spending["Per Student Budget"], bins, labels=school_spending_labels)

#Group new data by spending ranges per student and calculate the average
spending_group = reduced_overview_spending.groupby("Spending Ranges (Per Student)").mean()

#Format dataframe
spending_group["Average Math Score"] = spending_group["Average Math Score"].map("{:.2f}".format)
spending_group["Average Reading Score"] = spending_group["Average Reading Score"].map("{:.2f}".format)
spending_group["% Passing Math"] = (spending_group["% Passing Math"] * 100).map("{:.2f}".format)
spending_group["% Passing Reading"] = (spending_group["% Passing Reading"] * 100).map("{:.2f}".format)
spending_group["% Overall Passing"] = (spending_group["% Overall Passing"] * 100).map("{:.2f}".format)

spending_group
```
